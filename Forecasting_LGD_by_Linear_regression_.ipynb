#**Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

#**Reduced Scaled Features Dataset with 12 Predictors**

Importing Dataset

```python
data_int_scaled = pd.read_csv("/content/drive/MyDrive/TESI NPL's/data_int_scaled")
data_int_scaled = data_int_scaled.set_index("Pratica_Sequential")
data_int_scaled.head()
```

Splitting the dataset in predictors and target variable

```python
X_int_scaled = data_int_scaled.drop("LGD", axis=1)
y = data_int_scaled["LGD"]
```

```python
X_train_int_scaled, X_test_int_scaled, y_train_int_scaled, y_test_int_scaled = train_test_split(X_int_scaled, y, test_size=0.2, random_state=0)
```

Fitting Linear Regression

```python
Linear_Reg = LinearRegression()
```

```python
Linear_Reg.fit(X_train_int_scaled, y_train_int_scaled)
```

```python
y_pred_int_scaled = Linear_Reg.predict(X_test_int_scaled)
```

Estimating the evaluation metrics

```python
mae_int_scaled = mean_absolute_error(y_test_int_scaled, y_pred_int_scaled)
mse_int_scaled = mean_squared_error(y_test_int_scaled, y_pred_int_scaled)
rmse_int_scaled = np.sqrt(mse_int_scaled)
r2_int_scaled = r2_score(y_test_int_scaled, y_pred_int_scaled)
```

```python
print("MAE:", mae_int_scaled)
print("MSE:", mse_int_scaled)
print("RMSE:", rmse_int_scaled)
print("R2:", r2_int_scaled)
```
